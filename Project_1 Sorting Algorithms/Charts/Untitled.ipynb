```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pathlib import Path
```

```python
path = "../Results/"
```

```python
data0 = pd.read_csv(path + "heapsort.csv", sep = ";")
data1 = pd.read_csv(path + "introsort.csv", sep = ";")
data2 = pd.read_csv(path + "quicksort.csv", sep = ";")

fig, ax = plt.subplots(figsize=(20, 10))

ax.plot(data0["count"], data0["time"]/1000000, 'o-')
ax.plot(data1["count"], data1["time"]/1000000, 'o-')
ax.plot(data2["count"], data2["time"]/1000000, 'o-')

ax.grid();
ax.set(xticks = data0["count"]);

plt.xlabel("count")
plt.ylabel("time")
plt.show()
```
