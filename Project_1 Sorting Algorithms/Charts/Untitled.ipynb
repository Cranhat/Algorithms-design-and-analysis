```python
fig, ax = plt.subplots(figsize=(20, 10))
x = np.linspace(1, 5, 100)
ax.plot(x, [1 for i in x], '-', label = "O(1)", linewidth = 3)
ax.plot(x, x, '-', label = "O(n)", linewidth = 3)
ax.plot(x, np.log(x), '-', label = "O(log(n))", linewidth = 3)
ax.plot(x, x * np.log(x), '-', label = "O(nlog(n))", linewidth = 3)
ax.plot(x, x * x, '-', label = "O(n*n)", linewidth = 3)
ax.grid()

plt.legend(fontsize = "xx-large")
plt.show()
```
