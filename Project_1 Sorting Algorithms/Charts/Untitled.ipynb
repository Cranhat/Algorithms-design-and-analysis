```python
fig, ax = plt.subplots(figsize=(18, 12))
x = np.linspace(1, 5, 100)
ax.plot(x, [1 for i in x], '-', label = "O(1)", linewidth = 3)
ax.plot(x, [element if element > 1 else 1 for element in x], '-', label = "O(n)", linewidth = 3)
ax.plot(x, [np.log(element) if element > 1 else 0 for element in x], '-', label = "O(log(n))", linewidth = 3)
ax.plot(x,  [element * np.log(element) if element > 1 else 0 for element in x], '-', label = "O(nlog(n))", linewidth = 3)
ax.plot(x, x * x, '-', label = "O(n*n)", linewidth = 3)
ax.grid()

plt.xscale("log")
ax.set_title("Wykres teoretycznych złożoności obliczeniowych")
plt.legend(fontsize = "xx-large")
plt.show()
```
